```python
# ==========================
# Membaca Data CSV
# ==========================
import pandas as pd

# Ambil nama file yang diupload
nama_file = list(uploadedDaya.keys())[0]

# Membaca file CSV
dataDaya = pd.read_csv(nama_file, sep=';')

# Menampilkan 5 data pertama
print("Data berhasil dibaca:")
display(dataDaya.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Memastikan pandas diimpor

# Diasumsikan dataSuhu dan dataDaya sudah dimuat dari sel sebelumnya

try:
    # ** LANGKAH 1: Identifikasi nama kolom yang relevan **
    # Ganti 'NAMA_KOLOM_SUHU' dengan nama kolom aktual di dataSuhu
    # yang berisi data suhu yang relevan untuk perhitungan efisiensi Anda.
    # Ganti 'NAMA_KOLOM_DAYA' dengan nama kolom aktual di dataDaya
    # yang berisi data daya yang relevan untuk perhitungan efisiensi Anda.
    # Jika Anda memerlukan beberapa kolom (misalnya, suhu input/output, daya input/output),
    # identifikasi semua kolom tersebut.

    # --- GANTI 'NAMA_KOLOM_SUHU' DAN 'NAMA_KOLOM_DAYA' DENGAN NAMA KOLOM ACTUAL ANDA ---
    kolom_suhu_relevan = 'dataSuhu.head' # <--- **GANTI INI DENGAN NAMA KOLOM SUHU YANG BENAR**
    kolom_daya_relevan = 'dataDaya' # <--- **GANTI INI DENGAN NAMA KOLOM DAYA YANG BENAR**
    # -----------------------------------------------------------------------------

    # Periksa apakah kolom yang ditentukan ada di dataframe
    if kolom_suhu_relevan == 'NAMA_KOLOM_SUHU' or kolom_daya_relevan == 'NAMA_KOLOM_DAYA':
         print("PENTING: Silakan ganti 'NAMA_KOLOM_SUHU' dan 'NAMA_KOLOM_DAYA' dengan nama kolom aktual dari file CSV Anda.")
         print("Periksa output dari dataSuhu.head() dan dataDaya.head() untuk nama kolom yang benar.")
    elif kolom_suhu_relevan not in dataSuhu.columns:
        print(f"Error: Kolom '{kolom_suhu_relevan}' tidak ditemukan di dataSuhu.")
        print("Silakan periksa output dataSuhu.head() untuk nama kolom yang benar.")
    elif kolom_daya_relevan not in dataDaya.columns:
        print(f"Error: Kolom '{kolom_daya_relevan}' tidak ditemukan di dataDaya.")
        print("Silakan periksa output dataDaya.head() untuk nama kolom yang benar.")
    else:
        # ** LANGKAH 2: Pastikan keselarasan data **
        # Jika kedua dataframe memiliki kolom waktu atau pengenal umum,
        # sebaiknya gabungkan (merge) berdasarkan kolom tersebut untuk keselarasan yang akurat.
        # Jika mereka memiliki indeks yang sama dan mewakili titik data yang sesuai,
        # Anda dapat melanjutkan dengan pemeriksaan keselarasan indeks.

        if dataSuhu.index.equals(dataDaya.index):
            # ** LANGKAH 3: Lakukan perhitungan efisiensi Anda di sini **
            # Ganti baris berikut dengan rumus efisiensi aktual Anda
            # menggunakan kolom suhu dan daya yang ditentukan.
            # Perhatikan potensi pembagian dengan nol jika rumus Anda melibatkan pembagian.
            # Gunakan .replace(0, np.nan) pada penyebut jika perlu dan .dropna()
            # setelah perhitungan untuk menangani kasus ini.

            # --- MULAI DARI PERHITUNGAN EFISIENSI ANDA ---
            # Contoh: Perhitungan rasio sederhana (GANTI INI)
            # Pastikan kolom adalah numerik sebelum perhitungan
            dataSuhu[kolom_suhu_relevan] = pd.to_numeric(dataSuhu[kolom_suhu_relevan], errors='coerce')
            dataDaya[kolom_daya_relevan] = pd.to_numeric(dataDaya[kolom_daya_relevan], errors='coerce')

            # Hilangkan baris di mana konversi ke numerik gagal
            dataSuhu.dropna(subset=[kolom_suhu_relevan], inplace=True)
            dataDaya.dropna(subset=[kolom_daya_relevan], inplace=True)

            # Contoh: Rasio Daya terhadap Suhu * 100 (GANTI INI DENGAN RUMUS ANDA)
            # Tangani potensi nol pada penyebut
            denominator = dataSuhu[kolom_suhu_relevan].replace(0, np.nan)

            # Penting: Pastikan dataDaya dan dataSuhu tetap selaras berdasarkan indeks
            # setelah menghapus baris. Anda mungkin perlu menyelaraskan kembali atau
            # menggabungkan jika penghapusan baris mempengaruhi keselarasan indeks.
            # Jika Anda menggunakan merge di Langkah 2, perhitungan Anda akan berbeda.
            aligned_daya = dataDaya.reindex(denominator.index)[kolom_daya_relevan]

            # ** GANTI RUMUS EFISIENSI BERIKUT DENGAN RUMUS ANDA **
            # efisiensi = (aligned_daya / denominator) * 100
            # Misalnya, jika efisiensi adalah (output / input) * 100
            # Anda perlu mengidentifikasi kolom output dan input daya/suhu
            # Contoh sederhana (tetap berdasarkan asumsi Daya/Suhu):
            efisiensi = (aligned_daya / denominator) * 100

            efisiensi = efisiensi.dropna() # Hapus hasil NaN

            # --- AKHIR DARI PERHITUNGAN EFISIENSI ANDA ---

            if not efisiensi.empty:
                # ==========================
                # Membuat Grafik Batang Efisiensi
                # ==========================

                # Menggunakan index sebagai label x
                labels = efisiensi.index.astype(str)
                values = efisiensi.values

                plt.figure(figsize=(12, 6))
                plt.bar(labels, values, color='skyblue')
                plt.xlabel('Index Data') # Ganti dengan label yang relevan jika ada kolom waktu
                plt.ylabel('Efisiensi (%)')
                plt.title('Efisiensi Berdasarkan Data')
                plt.xticks(rotation=45, ha='right') # Rotasi label x agar tidak bertumpuk
                plt.tight_layout() # Menyesuaikan layout agar rapi
                plt.show()
            else:
                print("Setelah perhitungan, data efisiensi kosong atau semua nol.")
        else:
            print("Indeks data suhu dan daya tidak cocok. Tidak bisa menggabungkan berdasarkan index.")
            print("Pertimbangkan menggunakan kolom waktu untuk merge jika tersedia.")

except Exception as e:
    print(f"Terjadi kesalahan saat menghitung atau memplot efisiensi: {e}")
    print("Mohon periksa kembali nama kolom dan formula perhitungan efisiensi Anda.")
```
